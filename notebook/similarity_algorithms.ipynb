## 1. Similarity Algorithms

We calculate similarity between users based on 3 factors:
- **Mutual Friends** (Jaccard Similarity) - 40% Weight
- **Shared Skills** (Jaccard Similarity) - 40% Weight
- **Same Location** - 20% Weight

```python
def calculate_jaccard_similarity(set1, set2):
    if not set1 and not set2:
        return 0.0
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union > 0 else 0.0

def calculate_user_similarity(user1, user2):
    friends_sim = calculate_jaccard_similarity(set(user1.get("friends", [])), set(user2.get("friends", [])))
    skills_sim = calculate_jaccard_similarity(set(user1.get("skills", [])), set(user2.get("skills", [])))
    location_sim = 1.0 if user1.get("location") == user2.get("location") else 0.0
    
    return (friends_sim * 0.4) + (skills_sim * 0.4) + (location_sim * 0.2)

# Let's test it with two users from our data
user_a = data['users'][0]
user_b = data['users'][1]
score = calculate_user_similarity(user_a, user_b)
print(f"Similarity between {user_a['name']} and {user_b['name']} is {score:.4f}")
```
