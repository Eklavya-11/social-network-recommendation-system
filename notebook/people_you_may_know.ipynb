## 2. People You May Know

Recommends friends by calculating the similarity score against all non-friends, and returning the top matches.

```python
def people_you_may_know(user_id, data, top_n=5):
    users = data.get("users", [])
    user_map = {u["id"]: u for u in users}
    if user_id not in user_map: return []
        
    target_user = user_map[user_id]
    target_friends = set(target_user.get("friends", []))
    
    suggestions = []
    for u in users:
        if u["id"] == user_id or u["id"] in target_friends:
            continue
        sim_score = calculate_user_similarity(target_user, u)
        if sim_score > 0:
            suggestions.append((u, round(sim_score, 3)))
            
    suggestions.sort(key=lambda x: x[1], reverse=True)
    return suggestions[:top_n]

target = data['users'][0]
print(f"Friend recommendations for {target['name']}:")
recs = people_you_may_know(target['id'], data)
for u, s in recs:
    print(f"- {u['name']} (Score: {s})")
```
