## 3. Pages You Might Like

Uses Collaborative Filtering: we look at what similar users liked, and score pages based on those users' similarity to the target.

```python
def pages_you_might_like(user_id, data, top_n=5):
    users = data.get("users", [])
    pages = data.get("pages", [])
    user_map = {u["id"]: u for u in users}
    page_map = {p["id"]: p for p in pages}
    if user_id not in user_map: return []
        
    target_user = user_map[user_id]
    target_liked_pages = set(target_user.get("liked_pages", []))
    
    user_similarities = {u["id"]: calculate_user_similarity(target_user, u) for u in users if u["id"] != user_id}
    
    page_scores = {}
    for u in users:
        if u["id"] == user_id: continue
        sim = user_similarities[u["id"]]
        if sim <= 0: continue
            
        for pid in u.get("liked_pages", []):
            if pid not in target_liked_pages:
                page_scores[pid] = page_scores.get(pid, 0.0) + sim
                
    suggestions = [(page_map[pid], round(score, 3)) for pid, score in page_scores.items() if pid in page_map]
    suggestions.sort(key=lambda x: x[1], reverse=True)
    return suggestions[:top_n]

print(f"Page recommendations for {target['name']}:")
page_recs = pages_you_might_like(target['id'], data)
for p, s in page_recs:
    print(f"- {p['name']} (Score: {s})")
```
